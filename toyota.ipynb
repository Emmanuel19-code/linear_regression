```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score
```

```python
read_data = pd.read_csv("ToyotaCorolla.csv")
```

```python
read_data.drop("Id",axis=1,inplace=True)
```

```python
read_data
```

```python
ohe = OneHotEncoder()
```

```python
model= read_data["Model"]
model_encode = ohe.fit_transform(read_data[["Model"]]).toarray()
```

```python
read_data["Model"] = model_encode
```

```python
read_data
```

```python
fuel_encode = ohe.fit_transform(read_data[["Fuel_Type"]]).toarray()
```

```python
fuel_encode
```

```python
read_data["Fuel_Type"] = fuel_encode
```

```python
read_data
```

```python
features = read_data.drop(["Color","Price"],axis=1)
target = read_data["Price"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,random_state=100,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
regression = LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
prediction = regression.predict(x_test)
```

```python
r2_score(y_test,prediction)
```
